Importação de bibliotecas utilizadas na primeira aula

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

Load do arquivo com tratamento de encoding, separador, remoção de linhas de cabeçalho e rodapé e tratamento de números

```python
dados = pd.read_csv("/content/A070231189_28_143_208.csv", encoding = "latin_1", skiprows = 3, skipfooter = 12, sep = ";", thousands=".", decimal=",")
dados.head()
```

```python
dados.tail()
```

```python
dados.info()
```

```python
pd.options.display.float_format = "{:,.2f}".format
```

```python
dados.mean()
```

```python
dados['2008/Ago']
```

```python
dados['2008/Ago'].mean()
```

```python
ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind = "bar", figsize=(18, 7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor de procedimentos hospitalares processados em Agosto de 2018 por UF")
plt.show()
```

```python
dados = dados.replace("-", 0)
dados['2021/Ago'] = pd.to_numeric(dados['2021/Ago'])
dados = dados.sort_values(by=['2021/Ago'], ascending= False)
ax = dados.plot(x="Unidade da Federação", y="2021/Ago", kind = "bar", figsize=(18, 7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor de procedimentos hospitalares processados em Agosto de 2021 por UF")
plt.show()
```
